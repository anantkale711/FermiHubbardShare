```python
keys = list(data_dict.keys())
print(keys)
sz_name1= keys[0]
sz_name2= keys[1]
nn_name1=keys[4]
nn_name2=keys[5]
```

```python
Nu,Nd = (3,3)

fig1,ax1 = plt.subplots()
fig2,ax2 = plt.subplots()
fig3,ax3 = plt.subplots()

ax1.set_title(f"16 site ED, n={(Nu+Nd)/16:.3f}, GS overlap")
ax2.set_title(f"16 site ED, n={(Nu+Nd)/16:.3f}, Spin correlations")
ax3.set_title(f"16 site ED, n={(Nu+Nd)/16:.3f}, Density correlations")
count=0

for filename in os.listdir(direc):
    if not filename.endswith(".npz"):
        continue
    print(filename)

    filename2 = filename[:-4]
    flist =  filename2.split("_")
    if len(flist) == 11:
        N_up = int(flist[1])
        N_down = int(flist[3])
        if N_up != Nu or N_down != Nd:
            continue
        
        U = float(flist[5])
        imb_i = float(flist[9])
        imb_f = float(flist[10])
        # if imb_f !=0: continue
        with open(os.path.join(direc,filename), 'rb') as f:
            data = np.load(f, allow_pickle=True)
            # for key in data.files:
                # print(key)
            data_dict = dict(data)
        # plt.plot(t1, data_dict['overlap'])
        c = f"C{count}"
        ax1.plot(data_dict['times'], data_dict['overlap'], label=f'$t_x-t_y$=({imb_i},{imb_f})', color=c)
        
        ax2.plot(data_dict['times'], data_dict[sz_name1], label=f'Cx, $t_x-t_y$=({imb_i},{imb_f})',color=c, ls='-')
        ax2.plot(data_dict['times'], data_dict[sz_name2],label=f'Cy, $t_x-t_y$=({imb_i},{imb_f})', color=c, ls='--')

        n = (N_up+N_down)/16
        ax3.plot(data_dict['times'], data_dict[nn_name1]-n**2, 
                 label=f'Cx, $t_x-t_y$=({imb_i},{imb_f})', color=c, ls='-')
        ax3.plot(data_dict['times'], data_dict[nn_name2]-n**2, 
                 label=f'Cy, $t_x-t_y$=({imb_i},{imb_f})', color=c, ls='--')
        count += 1
ax1.set_ylabel("GS overlap")
ax1.set_xlabel('Time (1/t)')
ax1.legend(loc=2, bbox_to_anchor=(1,1))
ax1.set_ylim(-0.05,1.05)
ax1.grid()

ax2.legend(loc=2, bbox_to_anchor=(1,1))
ax2.grid()
ax2.set_ylabel("Connected SzSz correlations")
ax2.set_xlabel('Time (1/t)')


ax3.legend(loc=2, bbox_to_anchor=(1,1))
ax3.grid()
ax3.set_ylabel("Connected nn correlations")
ax3.set_xlabel('Time (1/t)')
plt.show()
```
